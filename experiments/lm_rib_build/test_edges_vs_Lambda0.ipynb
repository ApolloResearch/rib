```python
import torch
import matplotlib.pyplot as plt
```

```python
tiny = "/mnt/ssd-interp/stefan/rib/experiments/lm_rib_build/out/modular_arithmetic_jacobian_new_lambdas_1e0_rib_graph.pt"
small = "/mnt/ssd-interp/stefan/rib/experiments/lm_rib_build/out/modular_arithmetic_jacobian_new_lambdas_1e5_rib_graph.pt"
large = "/mnt/ssd-interp/stefan/rib/experiments/lm_rib_build/out/modular_arithmetic_jacobian_new_lambdas_1e9_rib_graph.pt"
```

```python
edges_small = torch.load(small)["edges"][0][1]
edges_tiny = torch.load(tiny)["edges"][0][1]
edges_large = torch.load(large)["edges"][0][1]
```

```python
fig, [ax3, ax1, ax2] = plt.subplots(1, 3, figsize=(10, 5))
fig.supxlabel("Edges with l+1 index > 20")
im = ax1.imshow((edges_small[20:]))
ax1.set_title("small")
cbar = fig.colorbar(im, ax=ax1)
im = ax2.imshow((edges_large[20:]))
ax2.set_title("large")
cbar = fig.colorbar(im, ax=ax2)
im = ax3.imshow((edges_tiny[20:]))
ax3.set_title("tiny")
cbar = fig.colorbar(im, ax=ax3)
plt.show()
```

```python
fig, [ax1, ax2, ax3] = plt.subplots(1, 3, constrained_layout=True)
fig.suptitle("Log10 of edges, without 0 row/col")
ax1.imshow(torch.log10(edges_tiny[1:, 1:]))
ax2.imshow(torch.log10(edges_small[1:, 1:]))
ax3.imshow(torch.log10(edges_large[1:, 1:]))
ax1.set_title("Lambda[0]=1e0")
ax2.set_title("Lambda[0]=1e5")
ax3.set_title("Lambda[0]=1e9")
plt.show()
```

```python
torch.testing.assert_close(edges_tiny[1:, 1:], edges_small[1:110, 1:])
```
